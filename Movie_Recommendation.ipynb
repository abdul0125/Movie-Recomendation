#Importing important dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import csv
```

##Data Collection and preprosessing

```python
movie_data = pd.read_csv('/content/movies.csv',encoding='utf-8')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

##Selecting relevant features for recommendation system

```python
#selecting the relevant features for recommendation
selected_features = ['genres','keywords','tagline','cast','director']
```

```python
#Replacing the null values with null string
for feature in selected_features:
  movie_data[feature]=movie_data[feature].fillna("")
```

```python
#Combining all the relevant features
combined_features = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
```

```python
#Converting textual data to numerical data
vectorizer = TfidfVectorizer()
```

```python
new_data = vectorizer.fit_transform(combined_features)
```

```python
print(new_data)
```

```python
similarity = cosine_similarity(new_data)
print(similarity)
```

```python
print(similarity.shape)
```

```python
#Getting a movie name from the user
movie = input("Enter the movie : ")
```

```python
list_movies = movie_data['title'].tolist()
```

```python
#Finding the close match for the movie given by the user
close_match = difflib.get_close_matches(movie,list_movies)
print(close_match)
```

```python
close_match = close_match[0]
```

```python
#Finding the index of movie with title

index_of_movie = movie_data[movie_data.title==close_match]['index'].values[0]
```

```python
print(index_of_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
```

```python
print(similarity_score)
```

```python
#sorting in order of similarity score
sorted_similar_movie = sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_similar_movie)
```
